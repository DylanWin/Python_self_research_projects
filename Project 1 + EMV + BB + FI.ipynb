```python
# Import necessary packages
import pandas as pd
import pandas_datareader as pdr
from pandas_datareader import data
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
import datetime
import numpy as np

# package required for QQ plot
import scipy.stats as stats
```

```python
# Period of stock data comparison ( economy crisis 2008 in specific)
start_date = '2007-01-01'
end_date = '2010-12-31'

#Stocks from Healthcare sector, Health Care Equipment sub industry
#(ABT) stock symbol of a healthcare sector in SP500 that has been through the 2008 recession
#(ABMD) another stock symbol of a healthcare sector in SP500 that has been through the 2008 recession for comparison
#(A) another stock symbol of a healthcare sector in SP500 that has been through the 2008 recession for comparison

#Stocks from Material sector
#(GLD) a stock symbol of material sector

ABT =data.DataReader('ABT', 'yahoo', start_date, end_date)
ABMD =data.DataReader('ABMD', 'yahoo', start_date, end_date)
A =data.DataReader('A', 'yahoo', start_date, end_date)
GLD =data.DataReader('GLD', 'yahoo', start_date, end_date)
```

```python
# Concatenate stock symbols datas with matching dates
symbol_data=[ABT, ABMD, A, GLD]
symbols=['A', 'ABMD', 'ABT', 'GLD']
all_data=pd.concat(symbol_data, keys=symbols, names=['Symbol','Date'])
```

## 2)

```python
# Prove that all 4 stocks exhibit weak stationary property
# % price change splitted into 3 intervals where mean and variance of each interval is compared
# stock sequence in 'ABT', 'ABMD', 'A', 'GLD'
for tick in symbols:
    x=daily_price_change[tick].values
    one_third_datapoint=round(len(x)/3)
    two_third_datapoint=(one_third_datapoint*2)

    data1, data2, data3=x[0:one_third_datapoint], x[one_third_datapoint:two_third_datapoint], x[two_third_datapoint:] 
    mean1, mean2, mean3=data1.mean(), data2.mean(), data3.mean()
    var1, var2, var3=data1.var(), data2.var(), data3.var()
    print(tick)
    print('mean1')
    print(mean1)
    print('mean2')
    print(mean2)
    print('mean3')
    print(mean3)

    print('var1')
    print(var1)
    print('var2')
    print(var2)
    print('var3')
    print(var3)
    print('\n')

    # mean and variace of stocks seem to remain vary within small range only
    # weak stationarity not certain until ACF correlogram seen
```

```python
# Plot ACF correlogram to determine if stock exhibit weak stationarity 
plt.clf()
plt.figure(figsize=(16,6))
plt.subplot(2,2,1)
plt.acorr(daily_price_change['ABT'])
plt.title('ACF correlogram of ABT returns')

plt.subplot(2,2,2)
plt.acorr(daily_price_change['ABMD'])
plt.title('ABMD correlogram of ABT returns')

plt.subplot(2,2,3)
plt.acorr(daily_price_change['A'])
plt.title('A correlogram of ABT returns')

plt.subplot(2,2,4)
plt.acorr(daily_price_change['GLD'])
plt.title('GLD correlogram of ABT returns')
plt.tight_layout()

# ACF decreases quickly for all stocks
# This implies that the  stocks have weak stationarity
```

## 3)

```python
# Plot moving average for windows=42, 252 for each stock
# which can then be used to display plots of dual moving average crossover technique in next plot

data_mvg_avg_42=all_data_pivoting.rolling(window=42, min_periods=1, center=False).mean()
data_mvg_avg_252=all_data_pivoting.rolling(window=252, min_periods=1, center=False).mean()

plt.clf()
plt.figure(figsize=(16,6))

for i in range(1,len(symbols)+1):
    plt.subplot(2,2,i)

    all_data_pivoting['42']=data_mvg_avg_42.iloc[:,i-1]
    all_data_pivoting['252']=data_mvg_avg_252.iloc[:,i-1]
    plt.plot(all_data_pivoting.iloc[:,i-1])
    plt.plot(all_data_pivoting['42'])
    plt.plot(all_data_pivoting['252'])
    plt.title('Moving average of {}'.format(all_data_pivoting.columns[i-1]))
    plt.legend(labels=['Adj Close', '42', '252'], loc=0)
    del all_data_pivoting['42']
    del all_data_pivoting['252']
plt.show()
plt.tight_layout()
```

```python
# Plot of dual moving average crossover technique example

# signal 'indicator' either being 1 for short_mvg_avg >long_mvg_avg 
# or 0 for otherwise

# signal 'decision' either being 1 for long stock, 
# -1 for short stock,
# or 0 for no action

signal=pd.DataFrame(index=all_data_pivoting.index)

plt.clf()

for i in range(len(symbols)):
    signal['indicator']=0.0
    signal['decision']=0.0
    
    signal['short_mvg_avg']=data_mvg_avg_42.iloc[:,i]
    signal['long_mvg_avg']=data_mvg_avg_252.iloc[:,i]
    signal['indicator'][42:]=np.where(signal['short_mvg_avg'][42:]
                                      >signal['long_mvg_avg'][42:],1.0,0.0)
    signal['decision']=signal['indicator'].diff()
    
    subplot_value=int(220+(i+1))
    fig = plt.figure(figsize=(14,6))
    ax1 = fig.add_subplot(subplot_value)

    all_data_pivoting.iloc[:,i].plot(ax=ax1, color='r')
    signal['short_mvg_avg'].plot(ax=ax1, lw=1.5)
    signal['long_mvg_avg'].plot(ax=ax1, lw=1.5)
    #plt.plot(signal['long_mvg_avg'])
    plt.title('Dual_mvg_avg_crsvr of {}'.format(all_data_pivoting.columns[i]))
    plt.legend(labels=['Adj Close', '42', '252'], loc=0)
    
    
    ax1.plot(signal.loc[signal.decision == 1.0].index, 
             signal.short_mvg_avg[signal.decision == 1.0], '^', markersize=8, color='k')
    
    ax1.plot(signal.loc[signal.decision == -1.0].index, 
             signal.short_mvg_avg[signal.decision == -1.0], '^', markersize=8, color='m')


plt.show()
plt.tight_layout()

```

### 4.1) Commodity Channel Index (CCI)

```python
# Examples of Commodity Channel Index based on 20 day mvg_avg
# Often used to identify cyclical turns across asset classes
# Formula being:
#   avg_price= (High + Low + Close)/3
#   CCI= (avg_price - mvg_avg of avg_price)/(0.015 * standard dev of avg_price)
# scales by 0.015 as inverse factor 
# +100 implies overbought, -100 implies oversold

all_stock_data=[A, ABMD, ABT, GLD]
CCI_data=[0,0,0,0]
day_lag=20

for i in range(len(symbols)):
    
    # Computing CCI of stocks
    stock_data=all_stock_data[i]
    avg_price=(stock_data['High']+stock_data['Close']+stock_data['Low'])/3
    CCI=pd.DataFrame((avg_price-avg_price.rolling(day_lag, min_periods=1, center=False).mean())/(0.015*avg_price.rolling(day_lag, min_periods=1, center=False).std()))
    CCI_data[i]=CCI
```

```python
# Example plot of A Close price and Commodity Channel Index based on 20 day mvg_avg
plt.clf()
plt.figure(figsize=(16,6))
plt.subplot(2,1,1).set_xticklabels([])
stock_data=all_stock_data[0]
plt.plot(stock_data['Close'], color='blue')
plt.title('Close price of {}'.format(all_data_pivoting.columns[0]))
plt.ylabel('Close Price')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(CCI_data[0], color='black')
plt.axhline(y=100, color='r')
plt.title('CCI of {}'.format(all_data_pivoting.columns[0]))
plt.ylabel('CCI values')
plt.grid(True)

plt.tight_layout()

# By CCI plot below, stock is oversold most of time as avg_price remains below
# 100 benchmark(red line) for most periods
```

```python
# Example plot of GLD Close price and Commodity Channel Index based on 20 day mvg_avg
plt.clf()
plt.figure(figsize=(16,6))
plt.subplot(2,1,1).set_xticklabels([])
stock_data=all_stock_data[0]
plt.plot(stock_data['Close'], color='blue')
plt.title('Close price of {}'.format(all_data_pivoting.columns[3]))
plt.ylabel('Close Price')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(CCI_data[1], color='black')
plt.axhline(y=100, color='r')
plt.title('CCI of {}'.format(all_data_pivoting.columns[3]))
plt.ylabel('CCI values')
plt.grid(True)

plt.tight_layout()

# By CCI plot below, stock is oversold most of time as avg_price remains below
# 100 benchmark(red line) for most periods
```

### 4.2) Ease of Movement (EVM)

```python
# Example of Ease of Movement with 14 day lag
# Indicate price rise/fall with stock volume taken into account
# + value(>0) asset price is rising with ease base on magnitude of value
# (higher price rise, lower volume => lesser selling pressure => price rise change is high)
# - value(<0) asset price is falling with ease
# Formula being:
#   price_move= ((current_high + current _low)/2)-((previous_high + previous_low)/2)
#   box_ratio= (volume/100,000,000)/(current_high - current_low)
#   EMV= price_move / box_ratio

EVM_mvg_avg=[0,0,0,0]

for i in range(len(symbols)):
    stock_data=all_stock_data[i]
    price_move=((stock_data['High']+stock_data['Low'])/2)-((stock_data['High'].shift(1)+stock_data['Low'].shift(1))/2)

    box_ratio=(stock_data['Volume']/100000000)/(stock_data['High']-stock_data['Low'])
    EVM=price_move/box_ratio
    EVM_mvg_avg[i]=EVM.rolling(window=14, min_periods=1, center=True).mean()
```

```python
# Example plot of A Close price and EVM based on 14 day mvg_avg
plt.clf()
plt.figure(figsize=(16,6))
plt.subplot(2,1,1).set_xticklabels([])
stock_data=all_stock_data[0]
plt.plot(stock_data['Close'], color='blue')
plt.title('Close price of {}'.format(all_data_pivoting.columns[0]))
plt.ylabel('Close Price')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(EVM_mvg_avg[1], color='black')
plt.title('EVM of {}'.format(all_data_pivoting.columns[0]))
plt.ylabel('EVM values')
plt.grid(True)

plt.tight_layout()

# By EVM plot below, EVM value fluctuates close to zero except during late 2008
# whch could be mostly affected by the economy crisis along with price drop of stock
```

```python
# Example plot of GLD Close price and EVM based on 14 day mvg_avg
plt.clf()
plt.figure(figsize=(16,6))
plt.subplot(2,1,1).set_xticklabels([])
stock_data=all_stock_data[0]
plt.plot(stock_data['Close'], color='blue')
plt.title('Close price of {}'.format(all_data_pivoting.columns[3]))
plt.ylabel('Close Price')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(EVM_mvg_avg[1], color='black')
plt.title('EVM of {}'.format(all_data_pivoting.columns[3]))
plt.ylabel('EVM values')
plt.grid(True)

plt.tight_layout()

# By EVM plot below, EVM value fluctuates close to zero except during late 2008
# whch could be mostly affected by the economy crisis along with price drop of stock
```

### 4.3) Bollinger Bands

```python
# Example of Bollinger Bands(BB) with 42 day lag
# Price is relatively high when price cross above upper BB(price might drop soon)
# Price is relatively low when price cross below lower BB(price might rise soon)
#  Upper BB= Closing_Price_mvg_avg
# Formula being:
#   upper_BB= simple_mvg_avg + (2 * std_dev(mvg_avg))
#   lower_BB= simple_mvg_avg - (2 * std_dev(mvg_avg))


upper_band_data=[0,0,0,0]
lower_band_data=[0,0,0,0]
mvg_avg_data=[0,0,0,0]
std_dev=[0,0,0,0]
lag=42

for i in range(len(symbols)):
    stock_data=all_stock_data[i]
    mvg_avg_data[i]=stock_data['Close'].rolling(window=lag, min_periods=1, center=False).mean()
    std_dev[i]=stock_data['Close'].rolling(window=lag, min_periods=1, center=False).std()
    lower_band_data[i]=mvg_avg_data[i]-(2*std_dev[i])
    upper_band_data[i]=mvg_avg_data[i]+(2*std_dev[i])
```

```python
# Example plot of A Close price and Bollinger Band based on 42 day mvg_avg
stock_data=all_stock_data[0]

plt.clf()
plt.figure(figsize=(16,7))
plt.plot(upper_band_data[0])
plt.plot(lower_band_data[0])
plt.plot(stock_data['Close'], color='r')
plt.title('BB plot of {}'.format(all_data_pivoting.columns[0]))
plt.legend(('Upper_Band', 'Lower_Band', 'Close'), loc=0)

# There were occasions where Closing price touches ipper/lower BB then remains around the BB
# for a period of time, thus not being a good technical indicator for this stock
```

```python
# Example plot of GLD Close price and Bollinger Band based on 42 day mvg_avg
stock_data=all_stock_data[3]

plt.clf()
plt.figure(figsize=(16,7))
plt.plot(upper_band_data[3])
plt.plot(lower_band_data[3])
plt.plot(stock_data['Close'], color='r')
plt.title('BB plot of {}'.format(all_data_pivoting.columns[3]))
plt.legend(('Upper_Band', 'Lower_Band', 'Close'), loc=0)

# There were many occasions where Closing price touches lower BB then proceed to touch upper BB
# thus being a better technical indicator for this stock(GLD) compared to previous one(A)
```

### 4.4) Force Index

```python
# Example of Force Index(FI) of period 15 days
# direction stock price, magnitude stock price movement, volume takken into account
# Determines the buying/selling pressure of assets
# Used to confirm trend or even predict upcoming reversals
# Formula being:
#   FI('period')=(current_close) - close(previous 'period')) * current_volume

FI=[0,0,0,0]
period=15

for i in range(len(symbols)):
    stock_data=all_stock_data[i]
    FI[i]=stock_data['Close'].diff(period)*stock_data['Volume']
```

```python
# Example of A Close price and Force Index of period 15 days
stock_data=all_stock_data[0]

plt.clf()
plt.figure(figsize=(16,7))
plt.subplot(2,1,1).set_xticklabels([])
plt.plot(FI[0])
plt.title('FI(15) of {}'.format(all_data_pivoting.columns[0]))
plt.subplot(2,1,2)
plt.plot(stock_data['Close'], color='r')
plt.title('Closing Price of {}'.format(all_data_pivoting.columns[0]))

# FI(15) seems to match reasonably well with the major closing price rise and fall of stock
# consider to be a good technical indicator for this stock
```

```python
# Example of GLD Close price and Force Index of period 15 days
stock_data=all_stock_data[3]

plt.clf()
plt.figure(figsize=(16,7))
plt.subplot(2,1,1).set_xticklabels([])
plt.plot(FI[3])
plt.title('FI(15) of {}'.format(all_data_pivoting.columns[3]))
plt.subplot(2,1,2)
plt.plot(stock_data['Close'], color='r')
plt.title('Closing Price of {}'.format(all_data_pivoting.columns[3]))

# FI(15) does not as match closing price of stock(GLD) well as previous stock(A)
# There could be many corrections in the given trend or even reversals
```
