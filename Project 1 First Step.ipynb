```python
# Motive is to observe if there a biggger possibility of higher correlations between stocks of same industry
# Observe if autocorrelations remain high during the 2008 economy crisis

import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot

#Stocks from Healthcare sector, Health Care Equipment sub industry
#(ABT) stock symbol of a healthcare sector in SP500 that has been through the 2008 recession
df_ABT=pd.read_csv(r'C:\Users\UX330\Documents\A Uni\Py\Python-for-Finance-Repo-master\csv stock price\ABT_2007_to_2009.csv',index_col='Date',parse_dates=True)

#(ABBV) another stock symbol of a healthcare sector in SP500 that has been through the 2008 recession for comparison
df_ABMD=pd.read_csv(r'C:\Users\UX330\Documents\A Uni\Py\Python-for-Finance-Repo-master\csv stock price\ABMD_2007_to_2009.csv',index_col='Date',parse_dates=True)

#(A) another stock symbol of a healthcare sector in SP500 that has been through the 2008 recession for comparison
df_A=pd.read_csv(r'C:\Users\UX330\Documents\A Uni\Py\Python-for-Finance-Repo-master\csv stock price\A_2007_to_2009.csv',index_col='Date',parse_dates=True)

#Stocks from Material sector
#(GLD) a stock symbol of material sector
df_GLD=pd.read_csv(r'C:\Users\UX330\Documents\A Uni\Py\Python-for-Finance-Repo-master\csv stock price\GLD_2007_to_2009.csv',index_col='Date',parse_dates=True)
```

```python
# Extracting daily returns of stocks
df_ABT_daily_returns=df_ABT['Adj Close'].pct_change()
df_ABMD_daily_returns=df_ABMD['Adj Close'].pct_change()
df_A_daily_returns=df_A['Adj Close'].pct_change()
df_GLD_daily_returns=df_GLD['Adj Close'].pct_change()

# # Extracting monthly returns of stocks
df_ABT_monthly_returns=df_ABT['Adj Close'].resample('M').ffill().pct_change()
df_ABMD_monthly_returns=df_ABMD['Adj Close'].resample('M').ffill().pct_change()
df_A_monthly_returns=df_A['Adj Close'].resample('M').ffill().pct_change()
df_GLD_monthly_returns=df_GLD['Adj Close'].resample('M').ffill().pct_change()
```

```python
# Plots to observe stocks daily returns during 2008 economy crisis period (btwn year 2007 - 2009)
plt.subplot(3,1,1)
df_ABT_daily_returns.plot(figsize=(16,6),label='ABT')
df_ABMD_daily_returns.plot(figsize=(16,6),label='ABMD')
df_A_daily_returns.plot(figsize=(16,6),label='A')
plt.legend(loc='upper left',fontsize=8)

plt.subplot(3,1,2)
df_GLD_daily_returns.plot(figsize=(16,6),label='GLD')
plt.legend(loc='upper left',fontsize=8)
plt.tight_layout()
```

```python
# Plots to observe stocks monthly returns during 2008 economy crisis period (btwn year 2007 - 2009)
plt.figure(figsize=(16,6))
plt.subplot(3,1,1)
df_ABT_monthly_returns.plot(figsize=(16,6),label='ABT')
df_ABMD_monthly_returns.plot(figsize=(16,6),label='ABMD')
df_A_monthly_returns.plot(figsize=(16,6),label='A')
plt.legend(loc='lower left',fontsize=10)

plt.subplot(3,1,2)
df_GLD_monthly_returns.plot(figsize=(16,6),label='GLD')
plt.legend(loc='upper left',fontsize=10)
```

```python
# Observe autocorrelation for max 1 month lag (assuming 30 days)
plt.figure(figsize=(16,6))
plt.subplot(2,2,1)
plt.acorr(df_ABT_daily_returns,maxlags=30)

plt.subplot(2,2,2)
plt.acorr(df_ABMD_daily_returns,maxlags=30)

plt.subplot(2,2,3)
plt.acorr(df_A_daily_returns,maxlags=30)

plt.subplot(2,2,4)
plt.acorr(df_GLD_daily_returns,maxlags=30)
# ?????
```

```python
# Observe autocorrelation for max 12 month lag (assuming 1 year)
plt.figure(figsize=(16,6))
plt.subplot(2,2,1)
plt.acorr(df_ABT_monthly_returns,maxlags=12)

plt.subplot(2,2,2)
plt.acorr(df_ABMD_monthly_returns,maxlags=12)

plt.subplot(2,2,3)
plt.acorr(df_A_monthly_returns,maxlags=12)

plt.subplot(2,2,4)
plt.acorr(df_GLD_monthly_returns,maxlags=12)
# ????
```
